```python
import xml.etree.ElementTree as ET

# Create a dummy XML file for demonstration
xml_data = '''
<movielist>
    <movie>
        <title>The Shawshank Redemption</title>
        <director>Frank Darabont</director>
        <year>1994</year>
    </movie>
    <movie>
        <title>The Godfather</title>
        <director>Francis Ford Coppola</director>
        <year>1972</year>
    </movie>
    <movie>
        <title>Pulp Fiction</title>
        <director>Quentin Tarantino</director>
        <year>1994</year>
    </movie>
</movielist>
'''

with open('movielist.xml', 'w') as f:
    f.write(xml_data)

print("Dummy 'movielist.xml' created successfully.")

# Parse the XML file
tree = ET.parse('movielist.xml')
root = tree.getroot()

# Iterate and print movie data
print("\nMovie DataBase")
for movie in root.findall("movie"):
    title = movie.find("title").text
    director = movie.find("director").text
    year = movie.find("year").text
    print(f"Title: {title}, Director: {director}, Year: {year}")
```

```python
import xml.etree.ElementTree as ET

# Create a dummy XML file for demonstration
xml_data ='''
<catalogo_peliculas>
    <pelicula>
        <titulo>Talentos Ocultos</titulo>
        <genero>Ficcion historica</genero>
        <anio>1960</anio>
        <director>Theodore Melfi</director>
        <pais_origen>Estados Unidos</pais_origen>
    </pelicula>
    <pelicula>
        <titulo>La vida es bella</titulo>
        <genero>Drama comedia</genero>
        <anio>1997</anio>
        <director>Roberto Benigni</director>
        <pais_origen>Italia</pais_origen>
    </pelicula>
    <pelicula>
        <titulo>Codigo enigma</titulo>
        <genero>Drama</genero>
        <anio>2014</anio>
        <director>Morten Tyldum</director>
        <pais_origen>Alemania</pais_origen>
    </pelicula>
</catalogo_peliculas>
'''

with open('catalogo_peliculas.xml', 'w') as f:
    f.write(xml_data)

print("Dummy 'catalogo_peliculas.xml' created successfully.")

# Parse the XML file
tree = ET.parse('catalogo_peliculas.xml')
root = tree.getroot()

# Iterate and print catalogo_peliculas data
print("\npelicula DataBase")
for pelicula in root.findall("pelicula"):
    titulo = pelicula.find("titulo").text
    genero = pelicula.find("genero").text
    director = pelicula.find("director").text
    anio = pelicula.find("anio").text
    print(f"titulo: {titulo}, director: {director}, anio: {anio}, genero: {genero}")

```

```python
import xml.etree.ElementTree as ET

# Create a dummy XML file for demonstration
xml_data = '''
<biblioteca>
    <libro>
        <autor>Antoine de Saint-Exupery</autor>
        <titulo>El principito</titulo>
        <fecha_publicacion>1943</fecha_publicacion>
    </libro>

    <libro>
        <autor>Gabriel García Márquez</autor>
        <titulo>Cien años de soledad</titulo>
        <fecha_publicacion>1967</fecha_publicacion>
    </libro>

    <libro>
        <autor>Homero</autor>
        <titulo>La Odisea</titulo>
        <fecha_publicacion>700 a.C.</fecha_publicacion>
    </libro>
</biblioteca>
'''

with open('biblioteca.xml', 'w') as f:
    f.write(xml_data)

print("Dummy 'biblioteca.xml' created successfully.")

# Parse the XML file
tree = ET.parse('biblioteca.xml')
root = tree.getroot()

# Iterate and print biblioteca data
print("\nbiblioteca DataBase")
for libro_element in root.findall("libro"):
    autor = libro_element.find("autor").text
    titulo = libro_element.find("titulo").text
    fecha_publicacion = libro_element.find("fecha_publicacion").text
    print(f"autor: {autor}, titulo: {titulo}, fecha_publicacion: {fecha_publicacion}")
```
